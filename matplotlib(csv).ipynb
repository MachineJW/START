```python
#   
import csv
import matplotlib.pyplot as plt
f=open('seoul.csv')
data=csv.reader(f)
next(data)
result=[] #   

for row in data:
    if row[4] != '':
        result.append(float(row[4]))

plt.figure(figsize=(20,5))# (,)
plt.plot(result,color='b')
plt.show()
f.close()
```

```python
# 97 ~ 22    
import csv
import matplotlib.pyplot as plt
f=open('seoul.csv')
data=csv.reader(f)
next(data)
result=[]
y=[]

for row in data:
    date=row[0].split('-')
    if date[1] == '01' and date[2] == '20' :
        y.append(date[0])
        result.append(float(row[4]))
        print(result)
f.close
plt.figure(figsize=(20,5)) #     .     
plt.plot(y,result,'r.-',label='max_temp',)
plt.legend(loc=4)
plt.show()
```

```python
#97~22 1  
import csv
import matplotlib.pyplot as plt

f=open('seoul.csv')
data=csv.reader(f)
next(data)

result=[]

for row in data:
    date=row[0].split('-')[1]
    if row[4] != ''and date=='01':
        result.append(float(row[4]))
plt.figure(figsize=(20,5))
plt.title('01_max_temp')      
plt.hist(result,bins=50,color='g',label='hist_test')
plt.legend()
plt.show
```

```python
#hist    
import csv
import matplotlib.pyplot as plt

f=open('seoul.csv')
data=csv.reader(f)
next(data)

result=[]

for row in data:
    month=row[0].split('-')[1]
    day=row[0].split('-')[2]
    date=month+day 
    
    if date=='0120':
        result.append(float(row[2]))

plt.title('my birthday')
plt.hist(result,bins=10,color='y',label='0120_temp')
plt.legend()
plt.show()
```

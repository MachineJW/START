```python
#   
import csv
import matplotlib.pyplot as plt
f=open('seoul.csv') # 97 ~ 22 02 
data=csv.reader(f)
next(data) #     
result=[] #   
print(next(data)) #     

for row in data:
    if row[4] != '':
        result.append(float(row[4]))

plt.figure(figsize=(20,5))# (,)
plt.plot(result,color='b')
plt.show()
f.close()
```

```python
# 97 ~ 22    
import csv
import matplotlib.pyplot as plt
f=open('seoul.csv')
data=csv.reader(f)
next(data)
result=[] #   
y=[] #   

for row in data:
    date=row[0].split('-') #    : date  row[0]   '-'  . (ex)1997-01-20 -> [1997,01,20]
    if date[1] == '01' and date[2] == '20' : #  date  2  01  3 20 (1 20)
        y.append(date[0]) #y date[0]<--  [1997  <---  date[0],01,20] 
        result.append(float(row[4])) #result      .
f.close
plt.figure(figsize=(20,5)) #     .     
plt.plot(y,result,'r.-',label='max_temp',)
plt.legend(loc=4)
plt.show()
```

```python
#97~22 1  
import csv
import matplotlib.pyplot as plt

f=open('seoul.csv')
data=csv.reader(f)
next(data)

result=[]

for row in data:
    date=row[0].split('-')[1] # date  row[0]    -   [1]   .    
    if row[4] != ''and date=='01': #     date 01 (<--- 1 ) 
        result.append(float(row[4])) #result   .
plt.figure(figsize=(20,5))#  
plt.title('01_max_temp')      
plt.hist(result,bins=50,color='g',label='hist_test')
plt.legend()
plt.show
```

```python
#hist    
import csv
import matplotlib.pyplot as plt

f=open('seoul.csv')
data=csv.reader(f)
next(data)

result=[]

for row in data:
    month=row[0].split('-')[1] #   
    day=row[0].split('-')[2] #   
    date=month+day  # date      
    
    if date=='0120': # date   0120  (120 )
        result.append(float(row[2])) # result    .

plt.title('my birthday')
plt.hist(result,bins=10,color='y',label='0120_temp')
plt.legend()
plt.show()
```
